```python
import numpy as np
import gym
import random
```

```python
# Create FrozenLake environment to train our RL agent
env = gym.make("FrozenLake-v0")
action_size = env.action_space.n
state_size = env.observation_space.n
Q = np.zeros((state_size, action_size))
#env.__dir__()
```

```python
# Hyperparameters for Q-learning algorithm
total_episodes = 15000
max_steps_per_episode = 100
alpha = 0.8 # Learning rate
gamma = 0.95 # Discount rate
# Exploration - Exploitation
initial_epsilon = 1.0
epsilon = 1.0
min_epsilon = 0.01
decay_rate = 0.005 # Exponential decay rate
```

```python
episode_rewards = []
for episode in range(total_episodes):
    # Reset environment
    state = env.reset()
    total_reward = 0;
    for step in range(max_steps_per_episode):
        # Choose an action given the current state
        # exploration Vs exploitation
        if random.random() < epsilon:
            # Exploration
            action = env.action_space.sample()
        else:
            # Exploitation
            action = np.argmax(Q[state, :])
                
        # Perform action
        observation, reward, done, info = env.step(action)
        
        # Update Qtable
        next_action = np.argmax(Q[observation, :])
        Q[state, action] = Q[state, action] + alpha * (reward + gamma * Q[observation, next_action] - Q[state, action])
        state = observation
        total_reward += reward
        
        # check if episode is done
        if done:
            break
    # Exploration - Exploitation exponential decay
    epsilon = max(min_epsilon, initial_epsilon * np.exp(-decay_rate * (episode + 1)))
    episode_rewards.append(total_reward)

print("Avg score (all episodes): {}".format(sum(episode_rewards) / total_episodes))
print("Avg score (last half episodes): {}".format(2 * sum(episode_rewards) / total_episodes))
# print("Rewards over time: {}".format(episode_rewards))
```

```python
# Play frozen lake with our agent
for episode in range(1):
    state = env.reset()
    total_reward = 0
    for step in range(max_steps_per_episode):
        env.render()
        action = np.argmax(Q[state,:])
        observation, reward, done, info = env.step(action)
        total_reward += reward
        state = observation
        if done:
            env.render()
            print("Total reward: {}".format(total_reward))
            print("Finished in: {} steps".format(step))
            break
```
